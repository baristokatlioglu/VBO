### Oyuncu İstatistikleri

* Oyuncunun o sezonki istatistikleri, toplam cinsinden
* 3258 farklı oyuncu
* 1979-80 ile 2021-22(43 sezon) arası
* 31 farklı takım (NBA'de 30 takım var. Verinin bazı noktalarında bir sezonda birden fazla takımda oynayan oyuncular için TOT(Two Other Teams) kullanılmış ve Takım ID'si 0 girilmiş.
O oyuncunun oynadığı takımlardaki istatistikleri veride mevcut. TEAM_ID'si 0 olanlar 2 takımda elde ettiği toplam istatistikler.

Glossary:
* GP: Game Played
* GS: Game Started
* MIN: Toplam oynanılan süre(dakika)
* FGM: Field Goal Made
* FGA: Field Goal Attempted
* FG_PCT: Field Goal Percentage
* FG3M: Field Goal 3 Made
* FG3A: Field Goal 3 Attempted
* FG3_PCT: Field Goal 3 Percentage
* FTM: Free Throw Made
* FTA: Free Throw Attempted
* FT_PCT: Free Throw Percentage
* OREB: Offensive Rebound
* DREB: Defensive Rebound
* REB: Rebound
* AST: Asist
* STL: Steal
* BLK: Block
* TOV: Turnover
* PF: Personal Foul
* PTS: Points
